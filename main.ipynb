```python
import os
import cv2
import numpy as np
```

```python

train_image_dir = "train/image"
train_mask_dir = "train/mask"
test_image_dir = "test/image"
test_mask_dir = "test/mask"

# --- Load training data ---
train_image_files = sorted([f for f in os.listdir(train_image_dir) if f.endswith('.png')])
train_mask_files = sorted([f for f in os.listdir(train_mask_dir) if f.endswith('.png')])

train_images = []
train_masks = []

for img_name, mask_name in zip(train_image_files, train_mask_files):
    img = cv2.imread(os.path.join(train_image_dir, img_name), cv2.IMREAD_COLOR)
    mask = cv2.imread(os.path.join(train_mask_dir, mask_name), cv2.IMREAD_GRAYSCALE)

    if img is None or mask is None:
        print(f"Could not load {img_name} or {mask_name}")
        continue

    train_images.append(img)
    train_masks.append(mask)

train_images = np.array(train_images)
train_masks = np.array(train_masks)

print("Loaded training data:")
print(f"Images shape: {train_images.shape}")
print(f"Masks shape:  {train_masks.shape}")

# --- Load test data ---
test_image_files = sorted([f for f in os.listdir(test_image_dir) if f.endswith('.png')])
test_mask_files = sorted([f for f in os.listdir(test_mask_dir) if f.endswith('.png')])

test_images = []
test_masks = []

for img_name, mask_name in zip(test_image_files, test_mask_files):
    img = cv2.imread(os.path.join(test_image_dir, img_name), cv2.IMREAD_COLOR)
    mask = cv2.imread(os.path.join(test_mask_dir, mask_name), cv2.IMREAD_GRAYSCALE)

    if img is None or mask is None:
        print(f"⚠️ Could not load {img_name} or {mask_name}")
        continue

    test_images.append(img)
    test_masks.append(mask)

test_images = np.array(test_images)
test_masks = np.array(test_masks)

print("Loaded test data:")
print(f"Images shape: {test_images.shape}")
print(f"Masks shape:  {test_masks.shape}")
```
